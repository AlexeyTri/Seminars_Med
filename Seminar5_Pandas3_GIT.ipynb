## Groupby Operations: Split-Apply-Combine

```python
df = pd.read_csv('/content/Seminars_Med/Seminars_Med/data/gapminder.csv')
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
avg_life_exp_by_year = df.groupby('year')['lifeExp'].mean()
avg_life_exp_by_year
```

```python
continent_describe = df.groupby('continent')["lifeExp"].describe()
continent_describe
```

```python
cont_le_agg = df.groupby('continent')['lifeExp'].agg(np.mean)
cont_le_agg
```

```python
def my_mean(values):
    n = len(values)
    sum = 0
    for value in values:
        sum += value
    return sum / n
```

```python
agg_my_mean = df.groupby('year')['lifeExp'].agg(my_mean)
agg_my_mean
```

```python
def my_mean_diff(values, diff_values):
    n = len(values)
    sum = 0
    for value in values:
        sum += value
    mean = sum / n
    return(mean - diff_values)
```

```python
global_mean = df['lifeExp'].mean()
```

```python
agg_mean_diff = (
    df.groupby('year')['lifeExp'].agg(my_mean_diff, diff_values=global_mean)
)
agg_mean_diff
```

```python
gdf = (
  df
  .groupby("year")
  ["lifeExp"]
  .agg([np.count_nonzero, np.mean, np.std])
)
gdf
```

```python
gdf_dict = df.groupby("year").agg(

  {
    "lifeExp": "mean",
    "pop": "median",
    "gdpPercap": "median"
  }
)
gdf_dict
```

```python
gdf = (
  df
  .groupby("year")
  ["lifeExp"]
  .agg(
    [
      np.count_nonzero,
      np.mean,
      np.std,
    ]
  )
  .rename(
    columns={
      "count_nonzero": "count",
      "mean": "avg",
      "std": "std_dev",
    }
  )
  .reset_index()
)
gdf
```

```python
import seaborn as sns
import numpy as np

# set the seed so results are deterministic
np.random.seed(42)

# sample 10 rows from tips
tips_10 = sns.load_dataset("tips").sample(10)

# randomly pick 4 'total_bill' values and turn them into missing
tips_10.loc[
    np.random.permutation(tips_10.index)[:4],
    "total_bill"
] = np.NaN
```

```python
tips_10
```

```python
count_sex = tips_10.groupby('sex').count()
count_sex
```

```python
def fill_na_mean(x):
  """Returns the average of a given vector"""
  avg = x.mean()
  return x.fillna(avg)
```

```python
# calculate a mean 'total_bill' by 'sex'
total_bill_group_mean = (
  tips_10
  .groupby("sex")
  .total_bill
  .transform(fill_na_mean)
)
```

```python
# assign to a new column in the original data
# you can also replace the original column by using 'total_bill'
tips_10["fill_total_bill"] = total_bill_group_mean
```
