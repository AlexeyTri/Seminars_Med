**Групповые и агрегированные расчеты**

* Для каждого года в наших данных, какова была средняя продолжительность жизни? Какова средняя продолжительность жизни, население и ВВП?

* А что, если мы рассортируем данные по континентам и выполним те же расчеты?

* Сколько стран перечислено на каждом континенте?

Чтобы ответить на только что поставленные вопросы, нам нужно выполнить *групповой (т. е. агрегированный)* расчет. Другими словами, нам нужно выполнить расчет, будь то средний или частотный подсчет, но применить его к каждому подмножеству переменной. Другой способ думать о групповых расчетах — это процесс *разделения-применения-объединения*. Сначала мы разделяем наши данные на различные части, затем применяем функцию (или расчет) по нашему выбору к каждой из разделенныхчасти, и, наконец, объединить все отдельные вычисления разделения в один фрейм данных. Мы выполняем групповые (т.е. агрегированные) вычисления, используя .*groupby()* метод на **DataFrames.**

```python
# For each year in our data, what was the average life expectancy?
# To answer this question, we need to:
# 1. split our data into parts by year
# 2. get the 'lifeExp' column
# 3. calculate the mean

df.groupby('year')['lifeExp'].mean()
```

```python
df_y = pd.DataFrame(df.groupby('year')['lifeExp'])
```

```python
df_y
```

```python
df_Y = pd.DataFrame(df.groupby(['year'])[['lifeExp']])
```

```python
df_Y
```

```python
df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean()
```

```python
df.loc[3]
```

```python
df_y.reset_index()
```

```python
df.groupby('continent')['country'].value_counts()
```

```python
df.groupby('continent')['country'].nunique()
```

```python
global_yearly_life_expectancy = df.groupby('year')['lifeExp'].mean()
```

```python
global_yearly_life_expectancy[:10]
```
